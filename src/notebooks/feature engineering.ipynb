```python
import sys
from sklearn.preprocessing import StandardScaler

sys.path.append('../features')
from data_cleaner import DataCleaner
from feature_engineer import FeatureEngineer
```

```python
cleaner = DataCleaner('../data/diem_thi_THPTQG_2026.csv')
df_raw = cleaner.load_data()
df_cleaned = cleaner.score_under_1()

engineer = FeatureEngineer(df_cleaned)
X_matrix = engineer.extract_features()

print(X_matrix.head(10))
```

```python
print(f'Executing feature normalization!')

scaler = StandardScaler()

feature_cols = ['F1_Personal_Std', 'F2_Z_Range', 'F3_Micro_Z', 'F4_Macro_Z']

X_scaled = X_matrix.copy()
X_scaled[feature_cols] = scaler.fit_transform(X_matrix[feature_cols])
print(X_scaled[feature_cols].describe().round(4))
```
